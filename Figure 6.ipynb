### Figure 3

```python
fig, axs= plt.subplots(nrows=3, ncols=1, figsize=(2.5,6.5), dpi=100, facecolor='white', constrained_layout=True)

### set font
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 8}
mpl.rc('font', **font)  # pass in the font dict as kwargs

### A ###
# set data
x = np.arange(0, 40, 1)
y7 = 1.152**x
y8 = 1.204**x
y9 = 1.282**x
axs[0].set_title('A', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
# combined effect
axs[0].plot(x, y9, label='more', color='#0072B2')
axs[0].plot(x, y8, label='medium', color='#E69F00')
axs[0].plot(x, y7, label='less', color='#999999')
axs[0].spines['right'].set_visible(False)
axs[0].spines['top'].set_visible(False)
axs[0].tick_params(axis='both', length=2, pad=1)
axs[0].set_xlabel('Slope (S)')
axs[0].set_ylabel('Physical Erosion (E)')
axs[0].annotate(r'$\mathbf{E \propto P}$', (30, 25))
axs[0].annotate(r'$\mathbf{E \propto S}$', (30, 10))
axs[0].set_ylim(bottom=0, top=200)
axs[0].set_xlim(left=0, right=40)
axs[0].legend(title='Precipitation (P):')._legend_box.align='left'
axs[0].set_xticks([])
axs[0].set_yticks([])

### B ###
x = np.arange(0,40,0.1)
y4 = 50 - 1.12**x
y5 = 30 - 1.12**x
y6 = 10 - 1.12**x
y4_1 = 50 - 1.15**x
y5_1 = 30 - 1.15**x
y6_1 = 10 - 1.15**x
axs[1].set_title('B', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
axs[1].plot(x, y4, label='more', color='#0072B2',)
axs[1].plot(x, y5, label='medium', color='#E69F00')
axs[1].plot(x, y6, label='less', color='#999999')
axs[1].set_ylim(bottom=0, top=100)
axs[1].set_xlim(left=-0, right=40)
axs[1].tick_params(axis='both', length=2, pad=1)
axs[1].legend(title='Vegetation cover (V):', ncol=1, loc='upper left')._legend_box.align='left'
axs[1].set_xlabel('Slope (S)')
axs[1].set_ylabel('Chemical Weathering (W)')
axs[1].annotate(r'$\mathbf{W \propto V}$', (30, 40))
axs[1].annotate(r'$\mathbf{W \propto 1/E}$', (30, 30))
axs[1].spines['right'].set_visible(False)
axs[1].spines['top'].set_visible(False)
axs[1].set_xticks([])
axs[1].set_yticks([])


# ### C ###
x = np.arange(0, 40, 0.1)
y7 = 1.152**x
y8 = 50 - 1.12**x
x1 = np.arange(0, math.log(50,1.12), 0.1)
y9 = 1.152**x1 + 50 - 1.12**x1
x2 = np.arange(math.log(50,1.12), 40, 0.1)
y10 = 1.152**x2
axs[2].set_title('C', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
# combined effect
axs[2].plot(x, y7, label='E', color='#0072B2')
axs[2].plot(x, y8, label='W', color='#E69F00')
axs[2].plot(x1, y9, color='black', linestyle=':', linewidth=2, label='D')
axs[2].plot(x2, y10,  color='black', linestyle=':', linewidth=2)
axs[2].vlines(x=24.9,ymin=-2, ymax=1.152**25, colors='red', linestyles=':', linewidth=1)
axs[2].legend(loc='upper left')._legend_box.align='left'
axs[2].set_ylim(bottom=0, top=200)
axs[2].set_xlim(left=0, right=40)
axs[2].tick_params(axis='both', length=2, pad=1)
axs[2].spines['right'].set_visible(False)
axs[2].spines['top'].set_visible(False)
axs[2].set_xlabel('Slope (S)')
axs[2].set_ylabel('Denudation (D)')
axs[2].annotate('D = E + W', (2,80), fontweight='bold')
axs[2].set_xticks([])
axs[2].set_yticks([])

# plt.savefig("Results/260309/Figure 6.svg", dpi=150)
```
